```python
import awswrangler as wr
import matplotlib.pyplot as plt
import seaborn as sns

print("Querying AWS Athena for deduplicated skills...")

# The ELT Approach: We deduplicate and flatten the data using cloud compute
query = """
WITH RankedJobs AS (
  SELECT 
    title, 
    company, 
    extracted_skills,
    ROW_NUMBER() OVER(PARTITION BY company, title ORDER BY created DESC) as rn
  FROM job_market_database.clean_jobs
),
UniqueJobs AS (
  SELECT title, company, extracted_skills
  FROM RankedJobs
  WHERE rn = 1
)
SELECT skill, COUNT(*) as demand_count
FROM UniqueJobs
CROSS JOIN UNNEST(extracted_skills) AS t(skill)
GROUP BY skill
ORDER BY demand_count DESC
LIMIT 15;
"""

# Fetch only the final aggregated counts into Pandas
df = wr.athena.read_sql_query(
    sql=query, 
    database="job_market_database"
)

print("Query complete! Building chart...")

# Build a clean, professional bar chart
plt.figure(figsize=(12, 7))
sns.barplot(data=df, x='demand_count', y='skill', hue='skill', palette='viridis', legend=False)

# Formatting the chart to look resume-ready
plt.title('Top 15 Most In-Demand Data Engineering Skills (2,000 Postings)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Number of Unique Job Postings', fontsize=12)
plt.ylabel('Skill', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the plot as an image for your GitHub README
plt.savefig('market_skills_chart.png', dpi=300)
plt.show()
```
